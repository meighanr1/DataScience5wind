```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 25)
y = np.linspace(-3, 3, 25)
X, Y = np.meshgrid(x, y)

plt.figure(figsize=(6,6))

for t in range(0, 10):
    plt.clf()
    
    # Time-varying wind field
    U = -Y + np.sin(t * 0.3)
    V = X + np.cos(t * 0.3)
    
    plt.quiver(X, Y, U, V)
    plt.title(f"Dynamic Wind Field (t={t})")
    plt.pause(0.3)

plt.show()
```
